**Theory**

In BI, raw data is usually at daily/campaign/product level.
We use SQL to summarize this data for reporting.

**Example business questions:**

- Total sales by campaign
- Total spends by date
- ROAS by category
- Campaigns with high spends but low sales
- Daily trend of orders

```sql
%sql
CREATE TABLE IF NOT EXISTS sales_data (
    date        DATE,
    campaign    STRING,
    category    STRING,
    spends      DOUBLE,
    sales       DOUBLE,
    orders      INT,
    clicks      INT,
    impressions INT
)
USING DELTA;
```

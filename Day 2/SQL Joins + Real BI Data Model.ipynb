**Practical Query**

**Join fact and campaign dimension**

```sql
SELECT
  c.campaign_name,
  SUM(f.spends) AS total_spends,
  SUM(f.sales) AS total_sales
FROM
  sales_fact f
    LEFT JOIN campaign_dim c
      ON f.campaign_id = c.campaign_id
GROUP BY
  c.campaign_name;
```

**Day 2 Practice**

1. Join sales_fact with product_dim

```sql
select
  p.product_id,
  sum(f.sales) as total_sales,
  sum(f.spends) as total_spends
from
  sales_fact f
    left join product_dim p
      on f.product_id = p.product_id
group by
  p.product_id;
```

2. Find total sales by category

```sql
select
  c.category,
  sum(f.sales) as total_sales
from
  sales_fact f
    left join campaign_dim c
      on f.campaign_id = c.campaign_id
group by
  c.category;
```

3. Find total spends by campaign_name

```sql
select
  c.campaign_name,
  sum(f.spends) as total_spends
from
  sales_fact f
    left join campaign_dim c
      on f.campaign_id = c.campaign_id
group by
  c.campaign_name;
```

4. Find products with  sales > 10000 using LEFT JOIN

```sql
select
  p.product_id,
  sum(f.sales) as no_sales
from
  sales_fact f
    left join product_dim p
      on f.product_id = p.product_id
group by
  p.product_id
having
  sum(f.sales) > 10000;
```

5. Explain INNER JOIN vs LEFT JOIN

INNER JOIN :
Returns only rows that have a match in both tables.

```sql
SELECT
  f.date,
  f.sales,
  c.campaign_name
FROM
  sales_fact f
    INNER JOIN campaign_dim c
      ON f.campaign_id = c.campaign_id;
```

Only returns rows where campaign_id exists in both sales_fact and campaign_dim. If a fact row has a campaign_id that doesn't exist in campaign_dim, it gets excluded.

LEFT JOIN :
Returns all rows from the left table, and matched rows from the right. If no match, right side returns NULL.

```sql
SELECT f.date, f.sales, c.campaign_name
FROM sales_fact f
LEFT JOIN campaign_dim c ON f.campaign_id = c.campaign_id;
```

Returns all rows from sales_fact even if campaign_id has no match in campaign_dim. Missing campaigns info shows as NULL.

| Scenario | Join to Use |
|---|---|
| Sales report — only show valid campaigns | `INNER JOIN` |
| Audit — find facts with missing campaign mapping | `LEFT JOIN` |
| Dashboard — show all dates even if no sales | `LEFT JOIN` on `date_dim` |
| Clean analytics on matched data only | `INNER JOIN` |
